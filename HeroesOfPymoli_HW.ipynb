```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
data_file = "../Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame

purchase_file_df = pd.read_csv(data_file)
purchase_file_df.head()
```

```python
#Player Count

#Display the total number of players.... THIS WORKS!!!
player_count = purchase_file_df.groupby("SN")["SN"].unique().count()
player_count
```

```python
#Purchasing Analysis (Total)
 
#Run basic calculations to obtain number of unique items, average price, etc.... THIS WORKS!!!
unique_items = len(purchase_file_df["Item ID"].unique())

average_price = purchase_file_df["Price"].mean()

no_of_purchases = purchase_file_df["Purchase ID"].count()

total_revenue = purchase_file_df["Price"].sum()


#Create a summary data frame to hold the results.... THIS WORKS!!!

summary_df = pd.DataFrame(dict(unique_items=unique_items, average_price=average_price, no_of_purchases=no_of_purchases, total_revenue=total_revenue),
                         index=[0]).round(2)
summary_df

#Optional: give the displayed data cleaner formatting

#Display the summary data frame
```

```python
#Gender Demographics.... THIS WORKS!!!

#Percentage and Count of Male Players
#Percentage and Count of Female Players
#Percentage and Count of Other / Non-Disclosed

gender_pct = ((purchase_file_df["Gender"].value_counts() / purchase_file_df["Gender"].count()) *100).round(2)
gender_pct

# #gender_count_female = purchase_file_df[purchase_file_df["Gender"]== "Female",].count()
# #gender_count_male = purchase_file_df[purchase_file_df["Gender"]== "Male",].count()

gender_count = purchase_file_df.groupby("Gender")["SN"].nunique()
gender_count

gender_table_df = pd.DataFrame({"Total Count": gender_count, "Percent of Players": gender_pct})
gender_table_df

#gender_groups = purchase_file_df.groupby("Gender").size()
#gender_groups

```

```python
#Purchasing Analysis (Gender)
 
#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender

#purchase count by gender..... THIS WORKS!!!
gender_purchase = purchase_file_df.groupby("Gender")["Item Name"].count()

#avg. purchase price by gender........ THIS WORKS!!!
gender_avg_purchase = purchase_file_df.groupby("Gender")["Price"].mean()

#avg. purchase total per person by gender.... THIS WORKS!!!
gender_avg_total = purchase_file_df.groupby("Gender")["Price"].sum()

average_pp = ((gender_avg_total)/(gender_count))

              
#Create a summary data frame to hold the results........ THIS WORK!!!
gender_summary_df = pd.DataFrame({"Purchase Count": gender_purchase, "Average Purchase Price": gender_avg_purchase, "Total Purchase Value": gender_avg_total, "Avg Total Purchase per Person": average_pp})
gender_summary_df["Average Purchase Price"]=gender_summary_df["Average Purchase Price"].map("${:.2f}".format)
gender_summary_df["Total Purchase Value"]=gender_summary_df["Total Purchase Value"].map("${:.2f}".format)
gender_summary_df["Avg Total Purchase per Person"]=gender_summary_df["Avg Total Purchase per Person"].map("${:.2f}".format)
gender_summary_df

#Optional: give the displayed data cleaner formatting

#Display the summary data frame
```

```python
#Age Demographics
 
#Establish bins for ages... THIS WORKS!!!

bins = [0,10,15,20,25,30,35,40,100]
age_catagory = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39","40+"]

## Categorize the existing players using the age bins. Hint: use pd.cut()...... THIS WORKS!!
# pd.cut(purchase_file_df["Age"], bins, labels=age_catagory).head()

purchase_file_df["Age Category"] = pd.cut(purchase_file_df["Age"], bins, labels=age_catagory)

# Calculate the numbers and percentages by age group.... THIS WORKS but numbers don't match
category_percent = ((purchase_file_df["Age Category"].value_counts() / player_count) *100).round(2)
#category_percent


## Create a summary data frame to hold the results
category_count = purchase_file_df["Age Category"].value_counts()

category_summary_df = pd.DataFrame({"Total Count": category_count, "Percentage of Players": category_percent})
category_summary_df

#Optional: round the percentage column to two decimal points

#Display Age Demographics Table
```

```python
#Purchasing Analysis (Age)
 
#Bin the purchase_data data frame by age
bins = [0,10,15,20,25,30,35,40,45]
age_catagory = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39",">=40"]

#Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below
group_count = purchase_file_df.groupby("Age Category")["Item Name"].count()
# group_count

group_avg = purchase_file_df.groupby("Age Category")["Price"].mean().round(2)
# group_avg

age_group_total = purchase_file_df.groupby("Age Category")["Price"].sum().round(2)
# age_group_total

group_average_pp = ((group_avg)/(group_count))

#Create a summary data frame to hold the results

group_summary_df = pd.DataFrame({"Purchase Count": group_count, "Average Purchase Price": group_avg, "Total Purchase Value": age_group_total, "Avg Total Purchase per Person": group_average_pp})
group_summary_df


#Optional: give the displayed data cleaner formatting

#Display the summary data frame
```

```python
#Top Spenders
 
#Run basic calculations to obtain the results in the table below
player_purchases = purchase_file_df.groupby("SN")["Price"].count()

avg_price = purchase_file_df.groupby("SN")["Price"].mean().round(2)

player_total = purchase_file_df.groupby("SN")["Price"].sum()

# decending=False

#Create a summary data frame to hold the results

player_summary_df = pd.DataFrame({"Purchase Count": player_purchases, "Average Purchase Price": avg_price, "Total Purchase Value": player_total})
# player_summary_df.head()

#Sort the total purchase value column in descending order
sorted_summary_df = player_summary_df.sort_values("Total Purchase Value", ascending=False)
sorted_summary_df

#Optional: give the displayed data cleaner formatting

#Display a preview of the summary data frame
```
